```python
all_codes = list(df['language_code'])
```

```python
## LID
gemma = []
gpt = []
gemini = []
for lang in all_codes:
    entry = lid[lid['language_code'] == lang]
    if entry.shape[0] > 0:
        gemma.append(entry.iloc[0]['Accuracy Gemma3-27B'])
        gpt.append(entry.iloc[0]['Accuracy gpt-4o'])
        gemini.append(entry.iloc[0]['Accuracy gemini2-Flash'])
    else:
        gemma.append('')
        gpt.append('')
        gemini.append('')
df['LID_accuracy_gemma'] = gemma
df['LID_accuracy_gpt4o'] = gpt
df['LID_accuracy_gemini'] = gemini
```

```python
## SIB
gemma = []
gpt = []
gemini = []
for lang in all_codes:
    entry = flores_sib[flores_sib['language_code'] == lang]
    if entry.shape[0] > 0:
        gemma.append(entry.iloc[0]['SIB Gemma-3-27B'])
        gpt.append(entry.iloc[0]['SIB gpt-4o (Apr 2025)'])
        gemini.append(entry.iloc[0]['SIB Gemini-2-Flash (Apr 2025)'])
    else:
        gemma.append('')
        gpt.append('')
        gemini.append('')
df['TC_accuracy_gemma'] = gemma
df['TC_accuracy_gpt4o'] = gpt
df['TC_accuracy_gemini'] = gemini
```

```python
## belebele
gemma = []
gpt = []
gemini = []
for lang in all_codes:
    entry = belebele[belebele['language_code'] == lang]
    if entry.shape[0] > 0:
        gemma.append(entry.iloc[0]['Accuracy Gemma3-27B'])
        gpt.append(entry.iloc[0]['Accuracy gpt-4o'])
        gemini.append(entry.iloc[0]['Accuracy gemini2-Flash'])
    else:
        if lang in inverted_belebele.keys():
            new_key = inverted_belebele[lang]
            entry = belebele[belebele['language_code'] == new_key]
            gemma.append(entry.iloc[0]['Accuracy Gemma3-27B'])
            gpt.append(entry.iloc[0]['Accuracy gpt-4o'])
            gemini.append(entry.iloc[0]['Accuracy gemini2-Flash'])
        else:  
            gemma.append('')
            gpt.append('')
            gemini.append('')
            
df['RC_accuracy_gemma'] = gemma
df['RC_accuracy_gpt4o'] = gpt
df['RC_accuracy_gemini'] = gemini
```

```python
## NLI
gemma = []
gpt = []
gemini = []
for lang in all_codes:
    if lang in inverted_nli.keys():
        nli_code = inverted_nli[lang]
        entry = nli[nli['language_code'] == nli_code]
        gemma.append(entry.iloc[0]['Accuracy Gemma3-27B'])
        gpt.append(entry.iloc[0]['Accuracy gpt-4o (April 2025)'])
        gemini.append(entry.iloc[0]['Accuracy Gemini-2-Flash (April 2025)'])
    elif lang in new_langs:
        entry = nli[nli['language_code'] == lang]
        gemma.append(entry.iloc[0]['Accuracy Gemma3-27B'])
        gpt.append(entry.iloc[0]['Accuracy gpt-4o (April 2025)'])
        gemini.append(entry.iloc[0]['Accuracy Gemini-2-Flash (April 2025)'])
    else:  
        gemma.append('')
        gpt.append('')
        gemini.append('')
            
df['NLI_accuracy_gemma'] = gemma
df['NLI_accuracy_gpt4o'] = gpt
df['NLI_accuracy_gemini'] = gemini
```

```python
### flores
gemma = []
gpt = []
gemini = []
for lang in all_codes:
    entry = flores_sib[flores_sib['language_code'] == lang]
    if entry.shape[0] > 0:
        gemma.append("{}/{}".format(entry.iloc[0]['Flores xx-eng Gemma3-27B'],entry.iloc[0]['Flores eng-xx Gemma3-27B']))
        gpt.append("{}/{}".format(entry.iloc[0]['Flores xx-eng gpt-4o (Apr 2025)'],entry.iloc[0]['Flores eng-xx gpt-4o (Apr 2025)']))
        gemini.append("{}/{}".format(entry.iloc[0]['Flores xx-eng Gemini-2-Flash (Apr 2025)'],entry.iloc[0]['Flores eng-xx Gemini-2-Flash (Apr 2025)']))
    else:
        gemma.append('')
        gpt.append('')
        gemini.append('')
df['MT_chrf_gemma'] = gemma
df['MT_chrf_gpt4o'] = gpt
df['MT_chrf_gemini'] = gemini
```

```python
df = df.drop(columns=["MT_chrf_gemma",'MT_chrf_gpt4o','MT_chrf_gemini'])
```
